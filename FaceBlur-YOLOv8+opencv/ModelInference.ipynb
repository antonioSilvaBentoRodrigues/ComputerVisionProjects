```python
yolo_model = YOLO('./runs/detect/train3/weights/best.pt')
```

```python
color = (190,144,16)
video_path = '/content/gdrive/MyDrive/FaceDetection/face_blur_sample.mp4'
output_path = '/content/gdrive/MyDrive/FaceDetection/op_face_blur_sample.mp4'
thresh=0.5
video_cap = cv2.VideoCapture(video_path)
ret, frame = video_cap.read()
H, W ,_ = frame.shape
output = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'MP4V'), int(video_cap.get(cv2.CAP_PROP_FPS)),(W,H))
while ret:
  results = yolo_model.track(frame, stream=True)
  for detection in results:
    classes_names = detection.names
    for bbox in detection.boxes:
      if bbox.conf[0] > thresh:
        [x1,y1,x2,y2] = bbox.xyxy[0]
        x1,y1,x2,y2 = int(x1), int(y1), int(x2), int(y2)
        class_ = int(bbox.cls[0])
        class_name = classes_names[class_]
        roi = frame[y1:y2, x1:x2]
        roi = cv2.GaussianBlur(roi, (0, 0), sigmaX=80)
        frame[y1:y2, x1:x2] = roi
        cv2.rectangle(frame, (x1,y1), (x2,y2), color, 3)
        cv2.putText(frame, f'{classes_names[int(class_)]} {bbox.conf[0]*100:.2f}%',
                    (x1,y1), cv2.FONT_HERSHEY_SIMPLEX,3,color,3)
  output.write(frame)
  ret, frame=video_cap.read()
video_cap.release()
output.release()
cv2.destroyAllWindows()
```
