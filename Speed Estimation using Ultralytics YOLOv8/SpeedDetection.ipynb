```python
from ultralytics import YOLO, solutions
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from google.colab import drive
```

```python
drive.mount('/content/gdrive/')
```

```python
model = YOLO('yolov8x.pt')
names = model.names
```

```python
video_path = '/content/gdrive/MyDrive/Projects/SpeedDetection/car_video_sample.mp4'
output_path = '/content/gdrive/MyDrive/Projects/SpeedDetection/car_video_sample_output.mp4'
video_cap = cv2.VideoCapture(video_path)
W, H, FPS = (int(video_cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
output = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*"mp4v"), FPS, (W,H))
```

```python
line_pts = [(5, int(H//2)), (int(W), int(H//2))]
```

```python
speed_obj = solutions.SpeedEstimator(
    names = names,
    reg_pts=line_pts,
    view_img=True)
```

```python
while video_cap.isOpened():
  ret, frame = video_cap.read()
  if not ret:
    print('Video frame empty')
    break
  results = model.track(frame, persist=True,show=False, conf=0.5)
  frame = speed_obj.estimate_speed(frame, results)
  output.write(frame)
video_cap.release()
output.release()
cv2.destroyAllWindows()
```
